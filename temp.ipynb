```python
import pandas as pd
import pyarrow.parquet as pq

# 文件路径
file_path = "/AI4M/users/qzh/lean_test/Agent/Temp/LeanRL/new_verl/verl/data/ds_prover/test.parquet"

# 方法 1：使用 pandas 读取整个数据集
df = pd.read_parquet(file_path)
print("数据集概览：")
print(df.head())  # 显示前几行
print("\n列名和数据类型：")
print(df.dtypes)  # 显示列名和数据类型
print("\n数据集形状：")
print(df.shape)   # 显示行数和列数

# 方法 2：使用 pyarrow 只查看元数据（不加载全部数据）
parquet_file = pq.ParquetFile(file_path)
print("\nParquet 文件元数据：")
print(parquet_file.metadata)  # 显示文件元数据
print("\n列名：")
print(parquet_file.schema)    # 显示数据集的 schema（列名和类型）
```
